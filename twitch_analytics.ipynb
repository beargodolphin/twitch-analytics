```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load game popularity table and create a dataframe for the top 20 games
gamepop_df = pd.read_csv("/Users/baileygodolphin/Desktop/Twitch/game_pop.csv")
gamepop_df = gamepop_df[:20]
```

```python
#Vizualise game popularity dataframe as bar chart
gamepop_df.plot.bar(x='game', y='count', rot=90)
plt.title('Viewer Count for Top 20 Featured Games on Twitch, January 1st 2015')
plt.xlabel('Games')
plt.ylabel('Viewers')
plt.legend(["Twitch"])
plt.show()
```

```python
#Load table for stream viewers per hour
timeofday_df = pd.read_csv("/Users/baileygodolphin/Desktop/Twitch/timeofday.csv")
```

```python
#Visualise viewers per hour in time series line graph
timeofday_df.plot.line(x='hour', y='count')
plt.title("Time Series for Stream Views per hour on Twitch, January 1st 2015")
plt.xlabel("Hour")
plt.ylabel("Viewers")
plt.legend(['01/01/2015'])
ax = plt.subplot()
ax.set_xticks(timeofday_df['hour'])
plt.show()
```
